```python
def emo_splitter(emotion):
    length=len(emotion)
    j=0
    emo=[[]]
    emo[0].append(emotion[0])
    emo[0].append(emotion[1])
    for i in range(2,length):
        emo[j].append(emotion[i])
        if emotion[i-2]!=emotion[i] and emotion[i]==emotion[i-1]:
            j+=1
            if i!= (length-1):
                emo.append([])
    return emo
```

```python
def act_splitter(emotion, action):
    length=len(emotion)
    j=0
    act=[[]]
    act[0].append(action[0])
    act[0].append(action[1])
    for i in range(2,length):
        act[j].append(action[i])
        if emotion[i-2]!=emotion[i] and emotion[i]==emotion[i-1]:
            j+=1
            if i!= (length-1):
                act.append([])
    return act
```

```python
def data_splitter(emotion, data):
    length=len(emotion)
    j=0
    dat=[[]]
    dat[0].append(data[0])
    dat[0].append(data[1])
    for i in range(2,length):
        dat[j].append(data[i])
        if emotion[i-2]!=emotion[i] and emotion[i]==emotion[i-1]:
            j+=1
            if i!= (length-1):
                dat.append([])
    return dat
```

```python
for i in range(len(old_utterance_emotion)):
    for l in emo_splitter(old_utterance_emotion[i]):
        utterance_emotion.append(l)
    for k in act_splitter(old_utterance_emotion[i], old_utterance_act[i]):
        utterance_act.append(k)
    for m in data_splitter(old_utterance_emotion[i], old_utterance_data[i]):
        utterance_data.append(m)
```

```python
utterance_corpus = {}
count = 0
for utterance in tqdm(utterance_data_new):
    
    utterance_info = [info.strip() for info in utterance.split(" __alp__")]
    
    if len(utterance_info) < 4:
        print(utterance_info)
        
    try:
        text, utt_idx, utt_act, utt_emo,speaker_id,conver_id = utterance_info[0], utterance_info[1], utterance_info[2], int(utterance_info[3]),utterance_info[4],utterance_info[5]
    except:
        print(utterance_info)
    
    meta = {'utt_emo': utt_emo, 'utt_idx': utt_idx,'utt_act': utt_act}
    
    # root & reply_to will be updated later, timestamp is not applicable 
    utterance_corpus[utt_idx] = Utterance(id=utt_idx, speaker= corpus_speakers[speaker_id], text=text, conversation_id=conver_id, meta= meta)

print("Total number of utterances = {}".format(len(utterance_corpus)))
```
