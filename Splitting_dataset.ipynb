```python
def emo_splitter(emotion):
    length=len(emotion)
    j=0
    emo=[[]]
    emo[0].append(emotion[0])
    emo[0].append(emotion[1])
    for i in range(2,length):
        emo[j].append(emotion[i])
        if emotion[i-2]!=emotion[i] and emotion[i]==emotion[i-1]:
            j+=1
            if i!= (length-1):
                emo.append([])
    return emo
```

```python
def act_splitter(emotion, action):
    length=len(emotion)
    j=0
    act=[[]]
    act[0].append(action[0])
    act[0].append(action[1])
    for i in range(2,length):
        act[j].append(action[i])
        if emotion[i-2]!=emotion[i] and emotion[i]==emotion[i-1]:
            j+=1
            if i!= (length-1):
                act.append([])
    return act
```

```python
def data_splitter(emotion, data):
    length=len(emotion)
    j=0
    dat=[[]]
    dat[0].append(data[0])
    dat[0].append(data[1])
    for i in range(2,length):
        dat[j].append(data[i])
        if emotion[i-2]!=emotion[i] and emotion[i]==emotion[i-1]:
            j+=1
            if i!= (length-1):
                dat.append([])
    return dat
```
